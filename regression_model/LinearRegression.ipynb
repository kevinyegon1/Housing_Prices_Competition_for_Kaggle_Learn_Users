```python
not_req_col = ['Id','MiscFeature','Fence','PoolQC','FireplaceQu','MasVnrType','Alley','SalePrice']
#train = train.drop(columns=['MiscFeature','Fence','PoolQC','FireplaceQu','MasVnrType','Alley'])
#test = train.drop(columns=['MiscFeature','Fence','PoolQC','FireplaceQu','MasVnrType','Alley'])
```

```python
# droped columns with more than 50% missing values
req_col =[]
for col in train.columns:
    if col not in not_req_col:
        req_col.append(col)
```

```python
# split dataset to features and target
X = train.drop(columns=['SalePrice'])
y = train['SalePrice']
test = test[req_col]
X = train[req_col]
```

```python
# choosing only categoric columns with 10 or less unique values for model to perform well
cat_col = X.select_dtypes('object').columns
req_cat_col = []
for col in cat_col:
    if X[col].nunique() <= 10:
        req_cat_col.append(col)
```

```python
num_col = X.select_dtypes(include=['int64','float64']).columns
req_col = list(num_col) + list(req_cat_col)
X = X[req_col]
test = test[req_col]
cat_col = X.select_dtypes(include=['object']).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LassoCV
```

```python
# filled nan with missing & encoded the categories using one hot encode
cat_processing = Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='constant', fill_value='Missing')),
        ('encoder',OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

```python
# filling nan values with median & scalling the figures for model to perform well
num_processing = Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)
```

```python
# wrapped both numeric and categoric transformation under the column Transformer
preprocessing = ColumnTransformer(
    transformers=[
        ('num', num_processing, num_col),
        ('cat', cat_processing, cat_col)
    ]
)
```

```python
import numpy as np

# log transforming the target to handle skewness
y = np.log1p(y)
```

```python
# created model inside the pipeline containing the preprocessing
model = Pipeline(
    steps=[
        ('preprocessing',preprocessing),
        ('reg',LassoCV(cv=5))
    ]
)
```

```python
# split the dataset to 80% training and 20% testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
# trained the model using X_train and y_train
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
# import evaluation techniques
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print(f'Mean Squared Error: {mean_squared_error(y_test,pred)}')
print(f'Mean Absolute Error: {mean_absolute_error(y_test,pred)}')
print(f'r2 Score: {r2_score(y_test,pred)}')
```
